```python
import pandas as pd
```

```python
df = pd.read_json('covid_data.json')

df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

df.head()
```

```python
df['YearMonth'] = df['date'].dt.to_period('M')

df.head()
```

```python
df_gb = df.groupby(['state', 'YearMonth']).sum()

df_gb.head()
```
