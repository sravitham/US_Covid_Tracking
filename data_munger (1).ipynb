```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('covid_data.json')

df.head()
```

```python
df.info()
```

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
df["positive"]= df["positive"].astype("Int64")
df["probableCases"]= df["probableCases"].astype("Int64")
df["negative"]= df["negative"].astype("Int64")
df["hospitalizedCumulative"]= df["hospitalizedCumulative"].astype("Int64")
df["recovered"]= df["recovered"].astype("Int64")
df["death"]= df["death"].astype("Int64")
df["positiveCasesViral"]= df["positiveCasesViral"].astype("Int64")
df["deathConfirmed"]= df["deathConfirmed"].astype("Int64")
# df['budget'] = df['budget'].astype("Int64") 
# df[['positive']]= df[['positive']].apply(pd.to_int)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

df.head()
```

```python
df['YearMonth'] = df['date'].dt.to_period('D')

df.head()
```

```python
df_gb1 = df.groupby([ 'YearMonth','state']).sum()

df_gb1.head(-10)
```

```python
df_gb1.reset_index().to_json("./Data/test1.json", orient='records')
```
